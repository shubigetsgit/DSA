```python
#Reverse a linked list in groups of given size

class Node:
    def __init__(self, val=None):
        self.val = val
        self.next = None

def reverse_groups(head, k):
    if head is None or k <= 1:
        return head
    prev = None
    curr = head
    next = None
    count = 0
    while curr is not None and count < k:
        next = curr.next
        curr.next = prev
        prev = curr
        curr = next
        count += 1
    if next is not None:
        head.next = reverse_groups(next, k)
    return prev
x = Node()
x.reverse_groups()
```

```python
#Merge a linked list into another linked list at alternate positions.
class Node:
    def __init__(self, val=None):
        self.val = val
        self.next = None

def merge_alternate(head1, head2):
    if head1 is None:
        return head2
    if head2 is None:
        return head1
    curr1 = head1
    curr2 = head2
    while curr1 is not None and curr2 is not None:
        next1 = curr1.next
        next2 = curr2.next
        curr1.next = curr2
        curr2.next = next1
        curr1 = next1
        curr2 = next2
    if curr2 is not None:
        curr1.next = curr2
    return head1
```

```python
# In an array, Count Pairs with given sum
def count_pairs(arr, target):
    count = 0
    seen = {}
    for num in arr:
        complement = target - num
        if complement in seen:
            count += seen[complement]
        seen[num] = seen.get(num, 0) + 1
    return count
```

```python
#Find duplicates in an array
def find_duplicates(arr):
    seen = set()
    duplicates = set()
    for num in arr:
        if num in seen:
            duplicates.add(num)
        else:
            seen.add(num)
    return list(duplicates)

```

```python
# Find the Kth largest and Kth smallest number in an array

def kth_largest(arr, k):
    arr.sort()
    return arr[-k]

def kth_smallest(arr, k):
    arr.sort()
    return arr[k-1]
```

```python
#Move all the negative elements to one side of the array
def move_negatives(arr):
    left = 0
    right = len(arr) - 1
    while left <= right:
        if arr[left] < 0:
            left += 1
        elif arr[right] >= 0:
            right -= 1
        else:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
    return arr
```

```python
#Reverse a string using a stack data structure

def reverse_string(s):
    stack = []
    for char in s:
        stack.append(char)
    reversed_str = ""
    while len(stack) > 0:
        reversed_str += stack.pop()
    return reversed_str
```

```python
# Evaluate a postfix expression using stack
def evaluate_postfix(expression):
    stack = []
    for char in expression:
        if char.isdigit():
            stack.append(int(char))
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()
            if char == '+':
                result = operand1 + operand2
            elif char == '-':
                result = operand1 - operand2
            elif char == '*':
                result = operand1 * operand2
            elif char == '/':
                result = operand1 / operand2
            stack.append(result)
    return stack.pop()

expression = '23+5*'
result = evaluate_postfix(expression)
print(result)
```

```python
#Implement a queue using the stack data structure
class Queue:
    def __init__(self):
        self.enqueue_stack = []
        self.dequeue_stack = []

    def enqueue(self, value):
        self.enqueue_stack.append(value)

    def dequeue(self):
        if not self.dequeue_stack:
            while self.enqueue_stack:
                self.dequeue_stack.append(self.enqueue_stack.pop())
        if not self.dequeue_stack:
            return None
        return self.dequeue_stack.pop()
    
    
queue = Queue()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
print(queue.dequeue()) 
print(queue.dequeue()) 
queue.enqueue(4)
print(queue.dequeue())
print(queue.dequeue())
```

```python
#Delete the elements in an linked list whose sum is equal to zero

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def delete_zero_sum_sublists(head):
    if not head:
        return None

    sum_map = {0: [head]}
    cumulative_sum = 0

    # Calculate the cumulative sum of each node and store them in a hash map
    current_node = head
    while current_node:
        cumulative_sum += current_node.value
        if cumulative_sum in sum_map:
            sum_map[cumulative_sum].append(current_node)
        else:
            sum_map[cumulative_sum] = [current_node]
        current_node = current_node.next

    # Remove nodes with the same sum
    current_node = head
    cumulative_sum = 0
    while current_node:
        cumulative_sum += current_node.value
        if len(sum_map[cumulative_sum]) > 1:
            prev_nodes = sum_map[cumulative_sum][:-1]
            for prev_node in prev_nodes:
                prev_node.next = None
            sum_map[cumulative_sum][0].next = sum_map[cumulative_sum][-1].next
            del sum_map[cumulative_sum]
        current_node = current_node.next

    # Remove nodes up to and including the node with a sum of zero
    dummy_node = Node(0)
    dummy_node.next = head
    current_node = dummy_node
    cumulative_sum = 0
    while current_node:
        cumulative_sum += current_node.value
        if cumulative_sum in sum_map:
            prev_nodes = sum_map[cumulative_sum][:]
            for prev_node in prev_nodes:
                prev_node.next = None
            current_node.next = sum_map[cumulative_sum][-1].next
            del sum_map[cumulative_sum]
        current_node = current_node.next

    return dummy_node.next
```
